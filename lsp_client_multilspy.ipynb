# LSP Client implemented with multilspy
`multilspy` is proposed in a research by microsoft, the [github repo](https://github.com/microsoft/monitors4codegen#4-multilspy)

```python
from monitors4codegen.multilspy import SyncLanguageServer
from monitors4codegen.multilspy.multilspy_config import MultilspyConfig
from monitors4codegen.multilspy.multilspy_logger import MultilspyLogger
```

```python
config = MultilspyConfig.from_dict(
    {"code_language": "rust"}
)  # Also supports "python", "rust", "csharp"
logger = MultilspyLogger()
lsp = SyncLanguageServer.create(config, logger, "/mnt/vm_data/pythonProjects/rust_analyzer_client/fd/")
with lsp.start_server():
    result = lsp.request_definition(
        "src/cli.rs",  # Filename of location where request is being made
        730,  # line number of symbol for which request is being made
        6,  # column number of symbol for which request is being made
    )
    
```

```python
result[0]
```

```python
with open(result[0]["absolutePath"]) as fp:
    lines = fp.readlines()
    start = result[0]["range"]["start"]["line"]
    end = result[0]["range"]["end"]["line"]
    target_lines = lines[start: end  + 1]
    content = "".join(target_lines)
    
print(content)
```
